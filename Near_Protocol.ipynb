```python
import pandas as pd
import plotly
import numpy as np
import plotly.express as px
```

```python
np_usd = pd.read_csv('/Users/shubhandua/Desktop/Pantera/PE Ratio Research/NP_USD.csv')
np_usd.head(10)
```

```python
np_usd['DateTime'] = np_usd['DateTime'].str.split("\s", n = 1, expand = True)[0]
np_usd.head(10)
```

```python
np_usd['Price']= np_usd['Price'].astype(float)
np_usd.dtypes
```

```python
np_usd.plot(x="DateTime", y= "Price",
        kind="line", figsize=(10, 10))
```

```python
np_usd['Price'] = np.log(np_usd['Price'])
```

```python
np_usd.plot(x="DateTime", y= "Price",
        kind="line", figsize=(15, 10))
```
